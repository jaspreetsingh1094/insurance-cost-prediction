# PERFORMING FEATURE ENGINEERING

```python
# Check unique values in the 'gender' column
insurance_df['gender'].unique()
```

```python
# convert categorical variable to numerical
insurance_df['gender'] = insurance_df['gender'].apply(lambda x: 0 if x == 'female' else 1)
```

```python
insurance_df.head()
```

```python
# Check the unique values in the 'smoker' column
insurance_df['smoker'].unique()
```

```python
# Convert categorical variable to numerical 
insurance_df['smoker'] = insurance_df['smoker'].apply(lambda x: 0 if x == 'no' else 1)
```

```python
insurance_df.head()
```

```python
# Check unique values in 'region' column
insurance_df['region'].unique()
```

```python
region_dummies = pd.get_dummies(insurance_df['region'], drop_first = True)
```

```python
region_dummies
```

```python
insurance_df = pd.concat([insurance_df, region_dummies], axis = 1)
```

```python
insurance_df.head()
```


# Finding which feature has the most positive correlation with charges?

```python
insurance_df.corr(numeric_only = True)
plt.figure(figsize = (10, 10))
sns.heatmap(insurance_df.corr(numeric_only = True), annot = True)
# smoker and age have positive correlations with charges
```

```python
# CREATING CATEGORIES OF CHARGES
insurance_df['charges_code'] = insurance_df['charges'].apply(lambda x: 1 if (x > 0 and x < 15000) 
                                                                        else (2 if (x > 15000 and x < 30000) 
                                                                        else (3 if (x > 30000 and x < 45000) 
                                                                        else 4))
                                                            )
```

```python
insurance_df
```

```python
# Dropping the original 'region' column 
insurance_df.drop(['region'], axis = 1, inplace = True)
insurance_df.head()
```

```python
# scaling the data before feeding the model
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler_x = StandardScaler()
X = scaler_x.fit_transform(X)

scaler_y = StandardScaler()
y = scaler_y.fit_transform(y)
```

```python
# Creating training and testing dataset with scaled data for linear regression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
print(X_train.shape)
print(X_test.shape)
```

# Train and evaluate a Linear Regression Model in Scikit-Learn

```python
# using linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, accuracy_score

regresssion_model_sklearn = LinearRegression()
regresssion_model_sklearn.fit(X_train, y_train)
```

```python
regresssion_model_sklearn_accuracy = regresssion_model_sklearn.score(X_test, y_test)
regresssion_model_sklearn_accuracy
```

```python
y_predict = regresssion_model_sklearn.predict(X_test)
```

```python
y_predict_orig = scaler_y.inverse_transform(y_predict)
y_test_orig = scaler_y.inverse_transform(y_test)
```

```python
k = X_test.shape[1]
n = len(X_test)
n
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt


RMSE = float(format(np.sqrt(mean_squared_error(y_test_orig, y_predict_orig)),'.3f'))
MSE = mean_squared_error(y_test_orig, y_predict_orig)
MAE = mean_absolute_error(y_test_orig, y_predict_orig)
r2 = r2_score(y_test_orig, y_predict_orig)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2) 
```
